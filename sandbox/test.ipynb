```python
from pymongo import MongoClient
import pandas as pd
```

```python
client = MongoClient('localhost', 27017)
db = client['sarscov2_standalone']
sample_coll = db.sample
variant_coll = db.variant
depth_coll = db.depth
depth = [(x) for x in depth_coll.find()]
sample = [(x) for x in sample_coll.find()]
```

```python
index = list()
for x in depth:
    for char in 'ATCG':
        if f'{x["pos"]}_{char}' not in index :index.append(f'{x["pos"]}_{char}')
df = pd.DataFrame(
    columns=[x['sample_id'] for x in sample],
    index=index
    )
appendList = list()
sample = 'sc2-1001'
for x  in depth:
    if x['sample_id'] == sample:
        appendList.append(x['A'])
        appendList.append(x['T'])
        appendList.append(x['C'])
        appendList.append(x['G'])
    else:
        df[sample] = appendList
        sample = x['sample_id']
        appendList = list()
        appendList.append(x['A'])
        appendList.append(x['T'])
        appendList.append(x['C'])
        appendList.append(x['G'])

```

```python
df.divide(df.max(axis=0), axis=1).transpose().fillna(df.divide(df.max(axis=0), axis=1).transpose().mean()).transpose()
```
